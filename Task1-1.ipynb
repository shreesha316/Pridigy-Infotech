Importing Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Reading the dataset

```python
df=pd.read_csv("C:/Users/Admin/OneDrive/Documents/Dataset/Population of India.csv")
df.head
data=pd.DataFrame(df)
data
```

Creating a Histogram for Continuous Data

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Population[50]', bins=30, kde=True)
plt.title('Distribution of Population in India')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Creating a Bar Chart for Categorial Data

```python
plt.figure(figsize=(14, 10))
sns.barplot(data=df, x='State/UT', y='Population[50]', palette='viridis')
plt.title('Population Distribution by State')
plt.xlabel('State')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.ylim(0, 219812380)
plt.show()
```

Gender-Wise Population

```python
# Extract relevant columns
gender_df = df[['State/UT', 'Male', 'Female']]

# Display the first few rows of the processed data
print(gender_df.head())

gender_df_melted = gender_df.melt(id_vars='State/UT', var_name='Gender', value_name='Population[50]')
```

Gender-Wise Population

```python
plt.figure(figsize=(14, 8))
sns.barplot(data=gender_df_melted, x='State/UT', y='Population[50]', hue='Gender', palette='viridis')
plt.title('Gender-wise Population Distribution by State')
plt.xlabel('State')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.ylim(0, 9.0 * 1e7)  # Set y-axis limits from 0 to 9.0 billion
plt.grid(True)
plt.show()
```
